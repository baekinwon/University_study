```python
import numpy as np

def l2_norm(x):
    return np.sqrt(np.sum(x**2))

def angle(x, y):
    # cos_theta = np.inner(x,y) / (l2_norm(x) * l2_norm(y))
    # theta = np.arccos(cos_theta)
    return np.arccos(np.inner(x,y)/(l2_norm(x)*l2_norm(y)))

a = np.array([1,0])
b = np.array([1,1])

print(angle(a,b))
```
