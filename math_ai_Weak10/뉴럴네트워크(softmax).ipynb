```python
import numpy as np

def softmax(z):
    denumerator = np.exp(z - np.max(z,axis=-1,keepdims=True)) # np.max를 빼는 이유는 오버플로우 때문임 소프트 맥스 식의 분자 식
    numerator = np.sum(denumerator,axis=-1,keepdims=True ) # softmax 식의 분모 식
    val = denumerator / numerator
    return val

z = np.array([[1,2,0],[1,1,1],[2,2,3]]) # 행렬
softmax(z) # 확률 값 구함
```
