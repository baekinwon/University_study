```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris = pd.DataFrame(data = iris.data, columns = iris.feature_names)
iris.sort_index(axis=0,ascending=False,inplace=True)
iris.head()
```

```python
iris.sort_index(axis=1, ascending = True, inplace=True)
iris.head()
```

```python
iris.sort_values('petal length (cm)',axis=0,ascending=False,inplace=True)
iris
```

```python
iris.sort_values('petal width (cm)',inplace=True)
iris
```

```python
import pandas as pd

hr1 = pd.DataFrame({'이름':['a','b'],
                   '부서':['영업','회계'],
                   '직급':['팀장','사원']})

hr2 = pd.DataFrame({'이름':['c','d'],
                   '부서':['영업','인사'],
                   '직급':['사원','팀장']})
```

```python
pd.concat([hr1,hr2],axis=0,ignore_index=False)
```

```python
pd.concat([hr1,hr2],axis=0,ignore_index=True)
```

```python
hr3 = pd.DataFrame({'이름':['e','f'],
                   '부서':['영업','인사'],
                   '급여':[3500,2800]})

pd.concat([hr1,hr3],axis=0,ignore_index=True)
```

```python
hr4 = pd.Series({0:2500},name='급여')
pd.concat([hr1,hr4],axis=1)
```

```python
hr5 = pd.DataFrame({'급여':[5000,3200,3500]})

pd.concat([hr1,hr5],axis=1)
```

```python
product = pd.DataFrame({'상품코드':['G1','G2','G3','G4'],
                       '상품명':['우유','감자','빵','치킨']})
product
```

```python
sale = pd.DataFrame({'주문번호':[1001,1002,1002,1003,1004],
                    '상품코드':['G4','G3','G1','G3','G5'],
                    '주문수량':[1,4,2,2,3]})
sale
```

```python
sale.merge(product, on = '상품코드',how='inner')
```
