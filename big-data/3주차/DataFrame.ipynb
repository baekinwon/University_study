```python
import pandas as pd
score = pd.DataFrame({'국어':[100,90],'영어':[80,75],
                     '수학':[80,95]},index=['철수','영희'])

score.loc['영희'] = 80
score
```

```python
score['국어']-=10
score
```

```python
score.loc['철수','영어'] = 90
score
```

```python
new_student = pd.Series({'국어':85,'영어':75,'수학':90},name='길동')
score = score.append(new_student)
score
```

```python
new_students = pd.DataFrame({'국어':[70,85],'영어':[65,100],
                     '수학':[95,65]},index=['영수','장화'])

score = score.append(new_students)
score
```

```python
score['사회'] = [85,70,95,80,70]
score
```

```python
score['사회'] += 5
score['총점'] = score['국어']+score['영어']+score['수학']+score['사회']
score
```

```python
score.drop('장화',inplace=True)
score
```

```python
score.drop(columns=['사회','총점'],inplace = True)
score
```

```python
score = pd.DataFrame({'이름':['a','b','c','d','e'],
                     '국어':[80,90,None,85,75],
                     '영어':[80,None,100,90,80],
                     '수학':[None,None,75,None,80]})

score[score['이름']=='b']
```

```python
score.loc[5,'이름':'수학'] = ['f',75,70,85]
score
```

```python
score.loc[(score['영어']>=80)&(score['수학']>=70),'합격']='Pass'
score
```

```python
score.loc[score['합격']!='Pass','합격']='Fail'
score
```

```python
import numpy as np

condition_list = [(score['국어']>=90),
                 (score['국어']>=80)&(score['국어']<90),
                 (score['국어']>=70)&(score['국어']<80)]

grade_list = ['A','B','C']

score['점수'] = np.select(condition_list,grade_list,default='F')
score
```
