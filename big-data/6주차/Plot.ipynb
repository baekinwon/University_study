```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

load_iris = load_iris()
load_iris
```

```python
iris = pd.DataFrame(load_iris.data,columns= load_iris.feature_names)
iris
```

```python
iris['class'] = load_iris.target
iris['class']
```

```python
iris['class'] = iris['class'].map({0:'setosa',1:'versicolor',
                                  2:'virginica'})
iris
```

```python
iris_type = iris['class'].value_counts()
iris_type
```

```python
plt.bar(x=iris_type.index,height=iris_type.values,width=0.8,align='center')
```

```python
plt.barh(y=iris_type.index,width=iris_type.values,
        height=0.5,align='edge')
```

```python
plt.hist('sepal width (cm)',bins=8,range=(1,6),color='purple',data=iris)
```

```python
plt.hist('sepal length (cm)',bins=5,range=(2,9),color='blue',data=iris)
```

```python
plt.boxplot(iris['sepal length (cm)','sepal width (cm)',''])
```

```python
plt.boxplot(iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']])
plt.show()
```

```python
iris[['sepal width (cm)','class']].boxplot(by='class')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x='class',y='sepal width (cm)',data=iris)
```

```python
def linear_func(x):
    return (2*x)+1

x = iris['sepal length (cm)']
plt.plot(x,linear_func(x),color='red')
```

```python
def func(x):
    return 3*x -2

x = iris['sepal width (cm)']
plt.plot(x,func(x),color='blue')
```

```python
import numpy as np
x,y = iris['sepal length (cm)'], iris['petal length (cm)']
b1, b0 = np.polyfit(x,y,1)

plt.scatter(x=x,y=y,alpha=0.5)

plt.plot(x,b1*x+b0,color='red')
```

```python
x,y = iris['sepal length (cm)'], iris['petal length (cm)']
b2, b1, b0 = np.polyfit(x,y,2)

plt.scatter(x=x,y=y,alpha=0.5)

plt.plot(x,(b2*x**2)+(b1*x)+b0,color='grey')
```

```python
iris2 = iris.sort_values(by='sepal length (cm)',axis= 0,ascending = True)

x,y = iris2['sepal length (cm)'], iris2['petal length (cm)']
b2, b1, b0 = np.polyfit(x,y,2)

plt.scatter(x=x,y=y,alpha=0.5)

plt.plot(x,(b2*x**2)+(b1*x)+b0,color='grey')
```
