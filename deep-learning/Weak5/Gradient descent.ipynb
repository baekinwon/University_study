```python
import numpy as np
```

```python
def func(x):
    return 0.5 * (x-2)**2 + 5
```

```python
import matplotlib.pyplot as plt

x1 = np.arange(-5,9,0.1)
# np.arange(최소값, 최대값, 간격): 최소값부터 최대값까지 지정된 간격으로 1D array 생성(최대값 포함 x)

x2 = np.linspace(-5,9,140)
# np.linspace(최소값, 최대값, 개수)
# 최소값부터 최대값까지 동일한 간격으로 지정된 개수만큼의 수를 1D array로 생성(최대값 포함)

# x1에 대한 line plot
plt.plot(x1,func(x1))
plt.grid()
```

```python
def grad_func(x):
    return x-2 # 위 func(x)의 return값을 미분한 값을 return 함
```

```python
iter = 20 # x 업데이트 횟수
eta = 0.3 # learning rate
x = -5 # 초기값
```

```python
for i in range(iter):
    print('[x]: ',x ,' [y]: ', func(x))
    x = x - eta * grad_func(x) # 경사하강법 식
```
