```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('C:/data/sample.csv').values
x = dataset[:,2]
x[6] = 55.0
y = dataset[:,3]
y[4] = 100000.0
```

```python
print(x)
print(y)
```

```python
np.arange(2,12,1)
# np.arange(a,b,c) : a부터 b까지 c 간격으로 1D array 생성
```

```python
np.linspace(2,12,10)
# np.linspace(a,b,c) : a부터 b까지 동일한 간격으로 c개의 value를 포함한 1D array 생성
# b포함
```

```python
# line plot : plt.plot(), 데이터가 하나일 경우 가로축은 데이터의 인덱스
plt.figure(figsize=(3,2)) # 차트 사이즈
axis_x = np.arange(200,1200,100)
plt.plot(axis_x,x,color='red',
         marker='>',
         markersize=10,
         linestyle='--',
         linewidth=3,
        label='chart 1')
# color는 선 색
# marker는 실제 데이터 있는 지점 표시
# linestyle: 선의 style
# linewidth : line 두께
# label : 한 figure에 여러 개의 차트 있을 때 각 차트의 이름 지정
plt.plot(axis_x,x+10,color='green',
         marker='s',
         markersize=10,
         linestyle='--',
         linewidth=3,
        label='chart 2')
plt.legend()
plt.title('line chart',fontsize=20,color='pink')
plt.xlabel('x')
plt.ylabel('age')
plt.grid() # 격자 표시
```

```python
plt.figure(figsize=(3,2))
plt.scatter(x,y,color='red',marker='s',s=50,label='data 1')
plt.scatter(x-5,y+20000,color='green',marker='o',s=50,label='data 2')
plt.legend()
# s는 marker의 사이즈
plt.grid()
plt.xlim([10,70]) # x축 범위 지정
plt.ylim([20000,150000]) # y축 범위 지정
```
